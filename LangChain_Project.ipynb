```python
prompt_formatting_s = SystemMessage(content=PROMPT_FORMATTING_S)
prompt_template_formatting_h = HumanMessagePromptTemplate.from_template(
    PROMPT_TEMPLATE_FORMATTING_H
)
chat_prompt_template_formatting = ChatPromptTemplate.from_messages([prompt_formatting_s, prompt_template_formatting_h])
```

### Create Embeddings, Vector Store, and Retriever

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embedding =OpenAIEmbeddings(
    model="text-embedding-3-small",  # Your Azure deployment name
    base_url="",
    api_key=""
)
```

```python
vectorstore = Chroma.from_documents(
    documents=docs_list_tokens_split,
    embedding=embedding,
    persist_directory="./intro-to-tableau"
)
```

```python
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 2, "lambda_mult": 0.7}
)
```

### Create Prompts and Prompt Templates for the Q&A Chatbot Chain

```python
PROMPT_CREATING_QUESTION = '''Lecture: {question_lecture}
Title: {question_title}
Body: {question_body}'''

PROMPT_RETRIEVING_S = '''You will receive a question from a student taking a Tableau course, which includes a title and a body. 
The corresponding lecture will also be provided.

Answer the question using only the provided context.

At the end of your response, include the section and lecture names where the context was drawn from, formatted as follows: 
Resources: 
Section: *Section Title*, Lecture: *Lecture Title* 
...
Replace *Section Title* and *Lecture Title* with the appropriate titles.'''

PROMPT_TEMPLATE_RETRIEVING_H = '''This is the question:
{question}

This is the context:
{context}'''

prompt_creating_question = PromptTemplate.from_template(PROMPT_CREATING_QUESTION)
prompt_retrieving_s = SystemMessage(content=PROMPT_RETRIEVING_S)
prompt_template_retrieving_h = HumanMessagePromptTemplate.from_template(PROMPT_TEMPLATE_RETRIEVING_H)

chat_prompt_template_retrieving = ChatPromptTemplate.from_messages([prompt_retrieving_s, prompt_template_retrieving_h])
```

### Create the First Version of the Q&A Chatbot Chain

```python
# Create the RunnableLambda function to extract text from StringPromptValue
RunnableLambda = lambda x: x.text

# Create the third component - a dictionary with context and question
# Create a function that retrieves context and formats the output
def retrieve_and_format(question_text):
    # Retrieve the two most relevant documents from the vector database
    docs = vectorstore.similarity_search(question_text, k=2)
    
    # Extract the page_content from each document
    context_docs = [doc.page_content for doc in docs]
    
    return {
        "context": context_docs,
        "question": question_text
    }
```

```python
# Build the chain with three components piped together
chain_retrieving = (
    prompt_creating_question
    | RunnableLambda
    | retrieve_and_format
    | chat_prompt_template_retrieving
    | chat
    | str_output_parser
)
```

```python
result = chain_retrieving.invoke({"question_lecture": "Adding a custom calculation",
                                  "question_title": "Why are we using SUM here? It's unclear to me.",
                                  "question_body": "This question refers to calculating the GM%."})
```

### Create a Runnable Function to Format the Context

```python
# Modified retrieve_and_format function - keep full Document objects
def retrieve_and_format(question_text):
    # Retrieve the two most relevant documents from the vector database
    docs = vectorstore.similarity_search(question_text, k=2)
    
    # Keep the full Document objects (not just page_content)
    return {
        "context": docs,  # Changed: keep full Document objects
        "question": question_text
    }
```

```python
# The format_context function stays the same
def format_context(dictionary):
    # Step 1: Define an empty string for formatted output
    formatted_string = ""
    
    # Step 2: Retrieve the list of Documents from dictionary
    retrieved_list = dictionary["context"]
    
    # Step 3: Loop through each Document object
    for i in range(len(retrieved_list)):
        # Append each document's content in the specified format
        formatted_string += f"""
Document {i}
Section Title: {retrieved_list[i].metadata.get('Section Title', '...')}
Lecture Title: {retrieved_list[i].metadata.get('Lecture Title', '...')}
Content: {retrieved_list[i].page_content}

- - - - - - - - - - - - - -
"""
    
    # Step 4: Create new dictionary with updated context
    new_dictionary = {
        "context": formatted_string,
        "question": dictionary["question"]
    }
    
    # Step 5: Return the new dictionary
    return new_dictionary
```

```python

# Step 6: Wrap the function in RunnableLambda and create the improved chain
chain_retrieving_improved = (
    prompt_creating_question
    | RunnableLambda
    | retrieve_and_format
    | format_context  # Add the new formatting function
    | chat_prompt_template_retrieving
    | chat
    | str_output_parser
)
```

```python
result_improved = chain_retrieving_improved.invoke({"question_lecture": "Adding a custom calculation",
                                                    "question_title": "Why are we using SUM here? It's unclear to me.",
                                                    "question_body": "This question refers to calculating the GM%."})
```

### Stream the Response

```python
result_streamed = chain_retrieving_improved.stream({
    "question_lecture": "Adding a custom calculation",
    "question_title": "Why are we using SUM here? It's unclear to me.",
    "question_body": "This question refers to calculating the GM%."
})

# Use a for-loop to print the generator's content consecutively
for chunk in result_streamed:
    print(chunk, end="", flush=True)
```
